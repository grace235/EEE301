```python

import numpy as np
import matplotlib.pyplot as plt
import math

import warnings
warnings.filterwarnings('ignore')
```

```python
voltage =  np.arange(158)
frequency = np.sqrt(voltage/(4* (np.pi**2) * (5 * 10**-9) * (1.6 * 10** -19)))
```

```python
plt.xlabel('Voltage')
plt.xlabel('Frequency')
plt.title('Plot Of Frequency Against Voltage')
plt.plot(voltage, frequency)
```
